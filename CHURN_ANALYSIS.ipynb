```python
# Count values and percentages
counts = df['Customer_Status'].value_counts()
pct = (df['Customer_Status'].value_counts() / len(df['Customer_Status']) * 100).round(3)

# Function to show both % and count
def func(pct, values):
    absolute = int(round(pct/100.*sum(values)))
    return "{:.1f}%\n({:d})".format(pct, absolute)

# Choosing a Seaborn color palette
palette = sns.color_palette("Set2", len(counts))

# Plot pie chart
plt.pie(
    counts.values,
    labels=counts.index,
    autopct=lambda pct: func(pct, counts.values),
    startangle = 90,
    colors = palette
)

plt.title("Customer Status Distribution")
plt.show()
```

> - This chart shows the customer status distribution. Around 67% of customers have stayed, while 26.5% have churned, and only 6.4% are new joiners. The big concern is that churn is much higher than the join rate, which could shrink customer base over time.”

```python
# Is Location Affecting the churning of customer ?

Top_churned_cities = (
    df[df["Customer_Status"] == "Churned"]   
      .groupby("City")
      .size()                                
      .sort_values(ascending=False)
      .head(10)
)
print (Top_churned_cities)
```

```python
# Plot bar chart
Top_churned_cities.plot(kind="bar", color="skyblue", figsize=(8,4))

plt.title("Top 10 Cities by Churned Customers", fontsize=14)
plt.xlabel("City", fontsize=12)
plt.ylabel("Number of Churned Customers", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

> - Most number of churn is coming from San Diego, followed by Los Angeles, while all other cities have much smaller numbers. This tells us that churn is heavily concentrated in just a couple of cities, and fixing issues there could make a big impact overall.

```python
# Creating countplots for all the categorical columns:

categorical_cols = [
    "Gender", "Married", "Offer", "Phone_Service", "Multiple_Lines",
    "Internet_Service", "Internet_Type", "Online_Security", 
    "Online_Backup", "Device_Protection_Plan", "Premium_Tech_Support",
    "Streaming_TV", "Streaming_Movies", "Streaming_Music", 
    "Unlimited_Data", "Contract", "Paperless_Billing", 
    "Payment_Method"
]

for col in categorical_cols:
    plt.figure(figsize=(8,4))
    ax = sns.countplot(data=df, x=col, hue="Customer_Status", palette="Set2")
    plt.title(f"Countplot of {col} by Customer_Status")
    plt.xticks(rotation=45)

    # Add counts INSIDE bars with dynamic font size
    for p in ax.patches:
        height = p.get_height()
        width = p.get_width()
        if height > 0:
            font_size = max(7, min(14, (height * width) ** 0.25))
            ax.annotate(f'{height}',
                        (p.get_x() + width / 2., height / 2),
                        ha='center', va='center',
                        fontsize=font_size, color='black', fontweight='bold')
            
            
    plt.show()
```


###  Insights:

#### 1. Contract Type
   > Customers on **month-to-month contracts churn the most** --> long-term contracts are the strongest loyalty driver.

#### 2. Internet Service & Type
   > Most churners had **internet service**, especially **fiber optic users** --> churn here is likely due to **pricing or service dissatisfaction**.

#### 3. Unlimited Data Plans
   > A majority of churners were on **unlimited data** --> the problem is **pricing and perceived value**, not the feature itself.

   > Customers without **online security, backup, device protection, or premium tech support** churn more --> value-added services create stickiness.

   > * **Bank withdrawal** users churn more --> they may be less engaged since payments are automatic.
   > * **Paperless billing** customers churn more --> digital-savvy users are price-sensitive and switch quickly.

#### 7. Customer Profile

   > * **Single/not married** customers churn more, while married customers are more loyal.
   > * **Gender has little effect** --> churn is almost equal between men and women.

#### 8. Offers & Plans

   > * Most churners had **no special offers** --> promotions and discounts could help retention.
   > * **Single-line users churn more** --> multi-line/family plans encourage loyalty.
   > * **Phone service** is widely used across all groups --> it’s a basic expectation, not a churn factor.

```python
# Exclude "Joined" customers
filtered_df = df[df["Customer_Status"] != "Joined"]

# Numerical columns to plot
num = ["Monthly_Charge", "Total_Charges", "Total_Refunds", 
       "Total_Extra_Data_Charges", "Total_Long_Distance_Charges", "Total_Revenue"]

# Set up subplots
fig, axes = plt.subplots(2, 3, figsize=(16, 12))
plt.subplots_adjust(wspace=0.3, hspace=0.5)
axes = axes.flatten()

# Custom palette for Churned vs Stayed
colors = ["#e4c87f", "#1b3780"]

for i, col in enumerate(num):
    sns.barplot(
        data=filtered_df,
        x="Customer_Status",
        y=col,
        hue="Customer_Status",   
        palette=colors,
        errorbar="ci",
        ax=axes[i],
        dodge=False
    )
    
    # Remove duplicate legends caused by hue
    axes[i].legend([],[], frameon=False)

    axes[i].set_xlabel("Customer Status", fontsize=11)
    axes[i].set_ylabel(f"Mean {col}", fontsize=11)
    axes[i].set_title(f"Average {col} by Customer Status", fontsize=13)
    axes[i].tick_params(axis="x", rotation=20)

# Hide unused subplot if fewer than 6
for i in range(len(num), 6):
    axes[i].axis("off")

plt.tight_layout()
plt.show()
```

### Insights:

#### 1. Monthly Charges
> Customers who churned had higher monthly charges, indicating that pricing is a major reason for leaving.

#### 2. Total Charges & Total Revenue
> Churned customers contributed much less to total charges and revenue, meaning they left before becoming valuable long-term.

#### 3. Refunds
> Customers who stayed received slightly more refunds/adjustments, while churned customers had fewer.
This suggests that not addressing billing or service issues may increase churn.

#### 4. Extra Data Charges
> Both churned and retained customers paid about the same, so extra data costs are not a churn factor.

#### 5. Long-Distance Charges
> Churned customers had much lower long-distance usage, meaning they were less engaged with the service and therefore more likely to leave.

## CONCLUSION:

Approximately 26.5% of clients have experienced churn in this data frame. According to this analysis, it is concluded that this churn has underlying causes, which are as follows:

### 1. Churn is concentrated in specific cities
> - San Diego and Los Angeles account for the highest churn, while other cities contribute much less.
Fixing issues in these cities could create a big overall impact.

### 2. Contract type strongly influences churn
> - Customers on month-to-month contracts churn the most, while long-term contracts significantly improve loyalty.

### 3. Internet service, especially Fiber Optic, drives churn
> - Most churners had internet service, and Fiber Optic users are leaving at higher rates, likely due to pricing or service dissatisfaction.

### 4. Early tenure churn dominates
> - 42% of churn happens within the first six months, with nearly half of these in the very first month.
By the end of one year, 55% of churn has already occurred, making the first year the most critical for retention.

### 5. Competitor switching is the biggest churn reason
> - Around 850 customers left for competitors, mainly due to better devices and better offers (~300 each).
> - Other competitor advantages like faster speeds or more data (~100–120 each) also play a role.

### 6. Customer service quality affects churn
> - ~310 customers left due to dissatisfaction or poor staff attitude, proving that service experience is as important as pricing.

### 7. Pricing pressure contributes to churn
> - Churned customers had higher monthly charges, but lower total charges and lifetime revenue (since they left early).
Price-related churn (~180–210 customers) is smaller than competitor churn but still significant.

### 8. Add-on services and bundles improve retention
> - Customers without online security, backup, device protection, or premium tech support churned more.
> - Similarly, customers without streaming bundles (TV, movies, music) were more likely to churn.
These features create stickiness and loyalty.

### 9. Customer profile influences churn
> - Single/not married customers churn more, while married customers are more loyal.
> - Gender has little effect, with churn rates nearly equal for men and women.
> - Multi-line/family plans reduce churn compared to single-line users.

### 10. Engagement and billing patterns matter
> - Customers with lower long-distance usage were more likely to churn, showing disengagement.
> - Bank withdrawal and paperless billing users churned more → digital-savvy, auto-payment customers may be more price-sensitive and quick to switch.
> - Refunds/adjustments were more common among retained customers, suggesting that addressing billing/service issues helps retention.
